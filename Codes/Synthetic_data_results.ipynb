```python
import pandas as pd
import numpy as np

def print_stat_details(data, metric):
    """Print detailed statistics for verification"""
    values = data[metric]
    print(f"\n{metric} Details:")
    print(f"All values: {sorted(values.unique())}")
    print(f"Min: {values.min():.3f}")
    print(f"Max: {values.max():.3f}")
    print(f"Mean: {values.mean():.3f}")
    print(f"Std: {values.std():.3f}")

def calculate_stats(df, algorithm):
    """Calculate mean and std for each metric for a given algorithm."""
    stats = {}
    # Get data for this algorithm
    data = df[df.index.get_level_values('algorithm') == algorithm]

    # Print raw data for verification
    print(f"\n{algorithm} Raw Data Verification:")
    print_stat_details(data, 'FDR')

    for metric in ['TPR', 'FDR', 'SHD']:
        values = data[metric]
        mean = values.mean()
        std = values.std()
        stats[metric] = {
            'mean': mean,
            'std': std,
            'formatted': f"{mean:.2f} ± {std:.2f}"
        }
    return stats

# Calculate for each algorithm
algorithms = ['DCD', 'PCMCI+', 'CD-NOD', 'CFCI', 'CPC', 'BOSS-LiNGAM']
all_stats = {}

for alg in algorithms:
    print(f"\n{'='*50}")
    print(f"Processing {alg}")
    print(f"{'='*50}")
    all_stats[alg] = calculate_stats(t1, alg)
    for metric in ['TPR', 'FDR', 'SHD']:
        stats = all_stats[alg][metric]
        print(f"{metric}: {stats['formatted']}")

# Create final summary table
columns = ['TPR ↑', 'FDR ↓', 'SHD ↓']
summary_data = {}

for alg in algorithms:
    alg_name = f"{alg} (Ours)" if alg == 'DCD' else alg
    summary_data[alg_name] = {
        'TPR ↑': all_stats[alg]['TPR']['formatted'],
        'FDR ↓': all_stats[alg]['FDR']['formatted'],
        'SHD ↓': all_stats[alg]['SHD']['formatted']
    }

summary_df = pd.DataFrame(summary_data).transpose()
summary_df = summary_df[columns]

print("\nFinal Summary Table:")
print(summary_df)
```

```python
import pandas as pd
import numpy as np

def format_mean_std(values):
    """Format mean ± std with 2 decimal places"""
    mean = values.mean()
    std = values.std()
    return f"{mean:.2f} ± {std:.2f}"

# Create empty lists to store results
rows = []
algorithms = ['DCD', 'PCMCI+', 'CD-NOD', 'CFCI', 'CPC', 'BOSS-LiNGAM']
metrics = ['TPR', 'FDR', 'SHD']
sample_sizes = [500, 1000, 1500]

# Calculate statistics for each algorithm and sample size
for algorithm in algorithms:
    row_data = []
    algorithm_name = f"{algorithm} (Ours)" if algorithm == "DCD" else algorithm

    for sample_size in sample_sizes:
        # Get data for this algorithm and sample size
        mask = (t1.index.get_level_values('algorithm') == algorithm) & \
               (t1.index.get_level_values('sample_size') == sample_size)
        data = t1[mask]

        # Calculate stats for each metric
        for metric in metrics:
            row_data.append(format_mean_std(data[metric]))

    rows.append([algorithm_name] + row_data)

# Create column names
column_names = []
for size in sample_sizes:
    for metric in metrics:
        column_names.append(f"{metric}")

# Create the DataFrame
df = pd.DataFrame(rows, columns=['Method'] + column_names)

# Add separating empty rows between lag groups
# First, determine how many algorithms are in each lag group
algorithms_per_group = len(algorithms)

# Create final DataFrame with spacing
final_rows = []
for i in range(0, len(rows), algorithms_per_group):
    # Add the group of algorithms
    final_rows.extend(rows[i:i+algorithms_per_group])
    # Add an empty row if this isn't the last group
    if i + algorithms_per_group < len(rows):
        final_rows.append(['' for _ in range(len(column_names) + 1)])

final_df = pd.DataFrame(final_rows, columns=['Method'] + column_names)

# Create a multi-level column index
column_tuples = [('Method', '')] + [(str(size), metric) for size in sample_sizes for metric in metrics]
final_df.columns = pd.MultiIndex.from_tuples(column_tuples)

# Print the table
print("\nTable: Statistics by Sample Size")
print("=" * 100)
print(final_df.to_string())

# Export to CSV and Excel
final_df.to_csv('statistics_by_sample_size.csv')
final_df.to_excel('statistics_by_sample_size.xlsx')
```
